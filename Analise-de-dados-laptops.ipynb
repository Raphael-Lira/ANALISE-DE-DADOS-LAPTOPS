```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('laptops.csv', sep=',')
display(df)
```

```python
df_laptops_clear = df[['name', 'price(in Rs.)', 'processor', 'ram', 'os', 'storage', 'display(in inch)', 'rating', 'no_of_ratings', 'no_of_reviews']]
display(df_laptops_clear)
```

```python
df_laptops_clear = df_laptops_clear.fillna(0)
df_laptops_clear.isna().sum()
```

```python
desc_preco = df_laptops_clear['price(in Rs.)'].describe()
print(desc_preco)
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data = df_laptops_clear, x = 'price(in Rs.)', kde = True,bins=50)
plt.title('Range de preco')
plt.show()
```

```python
lista_marcas = []
for laptop in df_laptops_clear['name']:
    lista_marcas.append(laptop.split()[0])
df_laptops_clear['Marca'] = lista_marcas
```

```python
marcas_principais = df_laptops_clear['Marca'].value_counts().head(6)
plt.figure(figsize=(12,6))
plt.bar(marcas_principais.index, marcas_principais.values, color=["#9219BC", "#208E31", "#2E43B0", "#E3D242", "#E3702F", '#E32F2F'])
plt.show()
```

```python
for processador, quantidade in df_laptops_clear['processor'].value_counts().items():
    if quantidade < 70:
        df_laptops_clear.loc[df_laptops_clear['processor'] == processador, 'processor']= 'Random'
df_laptops_clear['processor'].value_counts()
```

```python
plt.figure(figsize=(18,6))
df_laptops_clear.groupby('processor').size().sort_values(ascending = False).head(5).plot(kind = 'bar', color = sns.color_palette("coolwarm"))
plt.xlabel("Processor")
plt.ylabel("Frequency")
plt.title("Top 6 Processadores")
plt.xticks(rotation=0)
plt.show()
```

```python
for ram, quantidade in df_laptops_clear['ram'].value_counts().items():
    if quantidade < 23:
        df_laptops_clear.loc[df_laptops_clear['ram'] == ram, 'ram']= 'Random'
df_laptops_clear['ram'].value_counts()
```

```python
plt.figure(figsize=(12,6))
df_laptops_clear.groupby('ram').size().sort_values(ascending = False).head(6).plot(kind = 'bar', color = sns.color_palette("coolwarm"))
plt.xlabel("Variedades de ram")
plt.ylabel("Frequency")
plt.xticks(rotation=0)
plt.title("Top 6 RAMS")
plt.show()
```

```python
plt.figure(figsize=(15,6))
df_laptops_clear.groupby('display(in inch)').size().sort_values(ascending = False).head(6).plot(kind = 'bar', color = sns.color_palette("coolwarm"))
plt.xlabel("Tamanho de monitor")
plt.ylabel("Frequency")
plt.xticks(rotation=0)
plt.title("Top 6 tamanhos de monitores")
plt.show()
```
